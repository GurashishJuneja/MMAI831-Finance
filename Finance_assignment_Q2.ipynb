```
import pandas as pd

# There are 2 tables on the Wikipedia page, we want the first table
# second table shows the evolution of the changes

components = \
pd.read_html('https://en.wikipedia.org/wiki/List_of_S&P_500_companies')

first_table = components[0]
second_table = components[1]
sp_df = first_table
sp_df
```

```
SP500_PE=[]


for symbol in sp_df['Symbol']:
    try:
    #if symbol not in ['INFO', 'AES']:
        sym=symbol.replace('.', '-')  #For dual shares, eg, BRK.B, we have to change it to BRK-B
        ticker_yahoo = yf.Ticker(sym)
        print(symbol,round(ticker_yahoo.info['forwardPE'],2))
        SP500_PE.append(
            {
                "Ticker":sym,
                "trailingPE":ticker_yahoo.info['trailingPE'], 
                "forwardPE": ticker_yahoo.info['forwardPE'],
                "name": ticker_yahoo.info['longName']
            }
        )
    except:
        pass
    
SP500_PE=pd.DataFrame(SP500_PE)
```

```
SP500_PE
```

```
print(SP500_PE['trailingPE'].info())
```

```
# Convert 'trailingPE' column to numeric data type and drop any rows with missing values
SP500_PE['trailingPE'] = pd.to_numeric(SP500_PE['trailingPE'], errors='coerce')
SP500_PE = SP500_PE.dropna(subset=['trailingPE'])
#SP500_PE_no_inf = SP500_PE.replace([float('inf'), float('-inf')], " ").dropna(subset=['trailingPE'])

# Sort the DataFrame by descending order of 'trailingPE' values
SP500_PE_sorted = SP500_PE.sort_values('trailingPE', ascending=False)

# Get the 5 rows with the highest 'trailingPE' values
top_5_trailingPE = SP500_PE_sorted.nlargest(6, 'trailingPE')

# Round 'trailingPE' values to 2 decimal places
top_5_trailingPE['trailingPE'] = top_5_trailingPE['trailingPE'].round(2)
top_5_trailingPE['forwardPE'] = top_5_trailingPE['forwardPE'].round(2)

# Display the top 5 rows
print(top_5_trailingPE)
```

```
# Get the 5 rows with the lowest 'trailingPE' values
bottom_5_trailingPE = SP500_PE_sorted.nsmallest(6, 'trailingPE')

# Round 'trailingPE' values to 2 decimal places
bottom_5_trailingPE['trailingPE'] = bottom_5_trailingPE['trailingPE'].round(2)
bottom_5_trailingPE['forwardPE'] = bottom_5_trailingPE['forwardPE'].round(2)

# Display the bottom 5 rows
print(bottom_5_trailingPE)
```

```
# Convert 'forwardPE' column to numeric data type and drop any rows with missing values
SP500_PE['forwardPE'] = pd.to_numeric(SP500_PE['forwardPE'], errors='coerce')
SP500_FPE = SP500_PE.dropna(subset=['forwardPE'])

# Sort the DataFrame by descending order of 'forwardPE' values
SP500_PE_sorted_1 = SP500_FPE.sort_values('forwardPE', ascending=False)

# Get the 5 rows with the highest 'forwardPE' values
top_5_forwardPE = SP500_PE_sorted_1.nlargest(6, 'forwardPE')

# Round 'forwardPE' values to 2 decimal places
top_5_forwardPE['trailingPE'] = top_5_forwardPE['trailingPE'].round(2)
top_5_forwardPE['forwardPE'] = top_5_forwardPE['forwardPE'].round(2)

# Display the top 5 rows
print(top_5_forwardPE)
```

```
# Get the 5 rows with the lowest 'forwardPE' values
bottom_5_forwardPE = SP500_PE_sorted_1.nsmallest(6, 'forwardPE')

# Round 'forwardPE' values to 2 decimal places
bottom_5_forwardPE['trailingPE'] = bottom_5_forwardPE['trailingPE'].round(2)
bottom_5_forwardPE['forwardPE'] = bottom_5_forwardPE['forwardPE'].round(2)

# Display the bottom 5 rows
print(bottom_5_forwardPE)
```
